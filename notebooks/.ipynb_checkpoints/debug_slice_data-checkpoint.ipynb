```python
import numpy as np
import matplotlib.pyplot as plt
from jupyterthemes import jtplot
jtplot.reset()

total_iter = 25
datalen = 95
batch_size = 10
feat_dim = 16
x = np.zeros((datalen, feat_dim))
ticks = range(total_iter)

startlocpool = []
endlocpool = []
shapepool = []
for iter in ticks:
    start_idx = np.mod(iter*batch_size, datalen)
    end_idx = np.mod((iter + 1) * batch_size, datalen)
    startlocpool.append(start_idx)
    endlocpool.append(end_idx)

    x_batch = x[start_idx:end_idx] if end_idx > start_idx \
            else np.concatenate((x[start_idx:], x[:end_idx]), axis=0)

    
    
    shapepool.append(x_batch.shape)
    print(f'{x_batch.shape=}')
    print(f'{targloc=}')
    print(f'{targloc + batch_size=}')

uniqueshape = np.unique(np.array(shapepool), axis=0)
print(f'{uniqueshape=}')
assert (uniqueshape == np.array([batch_size, feat_dim])).all

plt.figure(figsize=(18, 10))
plt.plot(ticks, locpool)
plt.xlim([0, total_iter-1])
plt.ylim([0, datalen - batch_size])
plt.show()
```
