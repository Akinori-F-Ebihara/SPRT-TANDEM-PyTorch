```python
import sys, os, pdb
sys.path.append('/home/afe/Dropbox/PYTHON/SPRTproject/PytorchMAUS_dev/')
import numpy as np
from tqdm import tqdm
from termcolor import colored
from loguru import logger
from config.config import config
from utils.misc import grab_gpu, fix_random_seed, parse_args, extract_params_from_config
from utils.logging import create_log_folders, save_config_info, ContexualLogger,\
                          save_sdre_imgs, log_training_results, TensorboardLogger
from utils.hyperparameter_tuning import run_optuna, suggest_parameters_optuna, report_to_pruner
from utils.checkpoint import checkpoint_logger, update_and_save_result, \
                             initialize_objectives, finalize_objectives, load_pretrained_weights
from datasets.data_processing import SequentialDensityRatioData, slice_data 
from models.temporal_integrators import import_model
from models.optimizers import initialize_optimizer
from models.losses import compute_gradient, multiply_diff_mht
from utils.performance_metrics import llr_sequential_confmx, seqconfmx_to_macro_ave_sns,\
                                calc_llr_abserr, average_dicts
import pdb
import pyfiglet
from loguru import logger
import numpy as np
import pyfiglet
import torch
import torch.nn as nn
import torch.nn.functional as F
from utils.misc import extract_params_from_config

def import_model(config):
    '''
    Select LSTM- or TRANSFORMER- based temporal integrator.

    Args:
    - config (dict): dictionary that specifies model backbone.

    Returns:
    - model (tensorflow model)
    '''

    # check if necessary parameters are defined in the config file
    requirements = set(['MODEL_BACKBONE'])
    conf = extract_params_from_config(requirements, config)
    
    if any(char in conf.model_backbone.lower() \
            for char in ['lstm', 'long short-term memory', 'b2bsqrt_tandem']):
        model = B2BsqrtTANDEM(config)
    elif any(char in conf.model_backbone.lower() \
            for char in ['transformer', 'tfmr', 'tandemformer']):
        model = TANDEMformer(config)
    
    # print model layers and #parameters
    model.summarize_model()
    return model
    

def get_activation(activation):
    
    def B2Bsqrt(x, alpha=1.0):
        alpha=torch.tensor(alpha)
        return torch.sign(x) * (
            torch.sqrt(alpha + torch.abs(x)) - torch.sqrt(alpha)
        )
    
    def b2bsqrt_v2(x, alpha=1.0, beta=0.01):
        alpha=torch.tensor(alpha)
        beta=torch.tensor(beta)
        # add a linear component for anti-gradient vanishing
        return torch.sign(x) * (
            torch.sqrt(alpha + torch.abs(x))\
            + beta * torch.abs(x)\
            - torch.sqrt(alpha)
        )
    
    def b2bcbrt(x,  alpha=1.0, gamma=2.0):
        alpha=torch.tensor(alpha)
        gamma=torch.tensor(gamma)
        return torch.sign(x) * (
            torch.pow(alpha + torch.abs(x), 2/3)\
            - torch.pow(alpha, 2/3)
        )
    
    def b2bexp(x, alpha=0.03, beta=0.1, tau=1000):
        alpha=torch.tensor(alpha)
        beta=torch.tensor(beta)
        tau=torch.tensor(tau)
        return torch.sign(x) * (
            -alpha * torch.exp(-torch.abs(x) / tau)\
            + beta * torch.abs(x)\
            + alpha
        )
    
    def tanhplus(x, alpha=10.0, beta=0.02, tau=100.):
        alpha=torch.tensor(alpha)
        beta=torch.tensor(beta)
        tau=torch.tensor(tau)
        # add a linear component for anti-gradient vanishing
        return alpha * torch.tanh(x / tau) + beta * x
    
    def dullrelu(x, beta=0.05):
        beta=torch.tensor(beta)
        return beta * torch.maximum(torch.zeros_like(x), x)
    
    def B2Blog(x):
        return torch.sign(x) * (
            torch.log(1 + torch.abs(x))
        )
    
    if activation.lower() == 'b2bsqrt':
        return B2Bsqrt
    elif activation.lower() == 'b2bsqrtv2':
        return b2bsqrt_v2
    elif activation.lower() == 'b2bcbrt':
        return b2bcbrt
    elif activation.lower() == 'b2bexp':
        return b2bexp
    elif activation.lower() == 'tanhplus':
        return tanhplus
    elif activation.lower() == 'dullrelu':
        return dullrelu
    elif activation.lower() == 'b2blog':
        return B2Blog
    elif activation.lower() == 'relu':
        return nn.ReLU()
    elif activation.lower() == 'sigmoid':
        return nn.Sigmoid()
    elif activation.lower() == 'tanh':
        return nn.Tanh()
    else:
        raise ValueError(f'{activation} activation function is not supported!')


class B2BsqrtTANDEM(nn.Module):

    def __init__(self, config):
        
        super(B2BsqrtTANDEM, self).__init__()
        
        # check if necessary parameters are defined in the config file
        requirements = set(['NUM_CLASSES', 'WIDTH_LSTM', 'TIME_STEPS', 'FEAT_DIMS',
            'ACTIVATION', 'ACTIVATION_RECURRENT', 'ACTIVATION_FC', 'ACTIVATION_URGENCY',
            'IS_NORMALIZE', 'IS_POSITIONAL_ENCODING', 'IS_TRAINABLE_ENCODING', 
            'DROPOUT', 'ALPHA', 'BATCH_SIZE', 'IS_ADAPTIVE_LOSS', 'ORDER_SPRT'])
        conf = extract_params_from_config(requirements, config)

        # Parameters
        self.batch_size = conf.batch_size
        self.feat_dims = conf.feat_dims
        self.time_steps = conf.time_steps
        self.order_sprt = conf.order_sprt
        self.num_classes = conf.num_classes
        self.width_lstm = conf.width_lstm
        self.dropout = conf.dropout
        self.is_normalize = conf.is_normalize
        if self.is_normalize:
            self.layer_norm = nn.LayerNorm(
                normalized_shape=[self.feat_dims, self.time_steps]) 
        self.is_positional_encoding = conf.is_positional_encoding
        self.is_trainable_encoding = conf.is_trainable_encoding
        if self.is_trainable_encoding:
            self.trainable_pos_encoding = nn.Parameter(
                torch.randn(self.order_sprt + 1, self.feat_dims))

        # activation functions
        self.activation = get_activation(conf.activation)
        self.recurrent_activation = get_activation(conf.activation_recurrent)
        self.activation_logit = get_activation(conf.activation_fc)
        self.activation_urgency = get_activation(conf.activation_urgency)
        
        logger.warning('Currently B2Bsqrt is not supported.')
        self.rnn = nn.LSTM(
            input_size=self.feat_dims,
            hidden_size=self.width_lstm,
            num_layers=1,
            batch_first=True,
)
        # classification head: logits and urgency signals
        self.fc_logit = nn.Linear(in_features=self.width_lstm, out_features=self.num_classes, bias=True)
        self.fc_urgency = nn.Linear(in_features=self.width_lstm, out_features=1, bias=True)
        
        if conf.is_adaptive_loss:
            # adaptive loss weights
            self.adaptive_loss_weights = nn.Parameter(torch.randn(4))

    def features_to_logits(self, x, time_steps):
        """
        Args:
            x: A Tensor. Output of LSTM with shape=(batch, time_steps, self.width_lstm).
        Return:
            x: A Tensor. Logit with shape=(batch, time_steps, self.num_classes)        
        """
        logit_stack = []
        urgency_signals = []
        for i in range(time_steps):
            y = x[:, i, :]
            if self.is_normalize:
                y = self.layer_norm(y)
            
            # final FC layer
            y_class = self.activation_logit(y)
            y_class = self.fc_logit(y_class)
            logit_stack.append(y_class)
            
            # branch into the urgency signal
            y_urgency = self.activation_urgency(y)
            y_urgency = self.fc_urgency(y_urgency)
            urgency_signals.append(y_urgency)
        logit_stack = torch.stack(logit_stack, dim=1)
        urgency_signals = torch.stack(urgency_signals, dim=1) # shape(batch, time_steps, 1)
        return logit_stack, urgency_signals
    
    def positional_encoding(self, inputs, max_freq=10000):
        assert len(inputs.shape) == 3 # input size: (batch_size, time_steps, feat_dims)
        batch_size = inputs.shape[0]
        maxpos = inputs.shape[-2]
        d_model = inputs.shape[-1]
        x = np.arange(0, maxpos)
        y = np.arange(0, d_model)
        xx, yy = np.meshgrid(x, y, indexing='ij')
        arg = xx/max_freq**(2*(yy//2)/d_model)
        cos_mask = y % 2
        sin_mask = 1 - cos_mask
        pos_encoding = np.sin(arg) * sin_mask + np.cos(arg) * cos_mask
        return np.repeat(pos_encoding[np.newaxis, :, :], batch_size, dim=0)  

    def summarize_model(self, tb_writer):
        logger.info('\n' + pyfiglet.figlet_format(
            'SPRT-    \nTANDEM', font='banner3-D'))
        
        self.eval()
        example_input = torch.randn(self.batch_size, self.order_sprt + 1, self.feat_dims)
        with torch.no_grad():
            example_output, _ = self.forward(example_input)
            tb_writer.add_graph(self, example_input)

        # Print the model summary containing the number of trainable parameters
        num_params = sum(p.numel() for p in self.parameters() if p.requires_grad)
        logger.info(f'model summary:\n{self}')
        logger.info(f'input shape: {example_input.shape}')
        logger.info(f'output shape: {example_output.shape}')
        logger.info(f'Number of trainable parameters:, {num_params}')

    def forward(self, inputs):
        """Calc logits.
        Args:
            inputs: A Tensor with shape=(batch, time_steps, feature dimension). E.g. (128, 20, 784) for nosaic MNIST.
            training: A boolean. Training flag used in BatchNormalization and dropout.
        Returns:
            outputs: A Tensor with shape=(batch, time_steps, num_classes).
        """
        # inputs: (batch_size, time_steps, feat_dims)
        batch_size, time_steps, _ = inputs.shape 
       
        # Positinal encoding
        if self.is_positional_encoding:
            if self.is_trainable_encoding:
                inputs += self.trainable_pos_encoding.unsqueeze(0).repeat(batch_size, 1, 1)
            else:
                # (batch_size, time_steps, feat_dims)
                inputs += self.positional_encoding(inputs) 
        # Feedforward
        outputs, _ = self.rnn(inputs)

        # Make logits
        outputs, urgency_signal = self.features_to_logits(
            outputs, time_steps=time_steps)

        return outputs, urgency_signal # A Tensor with shape=(batch, order_sprt+1, num_classes)
    
```

```python
class TANDEMformer(nn.Module):
    '''
     Transformer with the Normalized Summation Pooling (NSP) layer for precise SDRE.
    '''
    def __init__(self, config):
        
        super(TANDEMformer, self).__init__()

        # check if necessary parameters are defined in the config file
        requirements = set(['NUM_CLASSES', 'TIME_STEPS', 'FEAT_DIMS',
            'NUM_BLOCKS', 'HEAD_SIZE', 'NUM_HEADS', 'FF_DIM', 'MLP_UNITS',
            'ACTIVATION', 'ACTIVATION_FC', 'ACTIVATION_URGENCY',
            'IS_NORMALIZE', 'IS_POSITIONAL_ENCODING', 'IS_TRAINABLE_ENCODING', 
            'DROPOUT', 'ALPHA', 'BATCH_SIZE', 'IS_ADAPTIVE_LOSS', 'ORDER_SPRT'])
        conf = extract_params_from_config(requirements, config)

        # Parameters
        self.batch_size = conf.batch_size
        self.time_steps = conf.time_steps
        self.order_sprt = conf.order_sprt
        self.feat_dims = conf.feat_dims
        self.num_transformer_blocks = conf.num_blocks
        self.is_normalize = conf.is_normalize
        self.is_positional_encoding = conf.is_positional_encoding # positional embedding
        self.is_trainable_encoding = conf.is_trainable_encoding

        # activation functions
        self.activation = get_activation(conf.activation)
        self.activation_fc = get_activation(conf.activation_fc)
        self.activation_urgency = get_activation(conf.activation_urgency)
        
        # transformer building blocks
        self.MultiHeadAttention = nn.MultiheadAttention(
            embed_dim=self.feat_dims, num_heads=conf.num_heads, dropout=conf.dropout)
        self.conv1 = nn.Conv1d(
            in_channels=self.feat_dims, out_channels=conf.ff_dim, kernel_size=1)
        self.conv2 = nn.Conv1d(
            in_channels=conf.ff_dim, out_channels=self.feat_dims, kernel_size=1)
        if self.is_normalize:
            self.layer_norm_1 = nn.LayerNorm(self.feat_dims)
            self.layer_norm_2 = nn.LayerNorm(self.feat_dims)
            self.layer_norm_3 = nn.LayerNorm(self.feat_dims)
        self.dropout = nn.Dropout(p=conf.dropout)

        # logit computation
        self.fc_transformer = nn.Linear(self.feat_dims, conf.mlp_units)
        self.fc_urgency = nn.Linear(conf.mlp_units, 1)
        self.fc_logits = nn.Linear(conf.mlp_units, conf.num_classes)
        if self.is_trainable_encoding:
            self.trainable_pos_encoding = nn.Parameter(
                torch.randn((self.order_sprt + 1, self.feat_dims)) * 0.1)

        if conf.is_adaptive_loss:
            # adaptive loss weights for mce, lllr, ausat, and weight decay
            self.adaptive_loss_weights = nn.Parameter(torch.ones(4))
            
    # transformer block 
    def transformer_encoder(self, inputs):
        # self-attention 
        x = inputs # (batch_size, time_steps, feat_dims)
        if self.is_normalize:
            x = self.layer_norm_1(x)
        # MultiHeadAttention returns the tuple of two tensors, 
        # output of multi-head attention and attention weights.
        x = self.MultiHeadAttention(x, x, x)[0]
        x = self.dropout(x)
        res = x + inputs

        # feedforward
        if self.is_normalize:
            x = self.layer_norm_2(res)
        else:
            x = res
        # PyTorch conv1d expect an input shape to be (batch_size, feat_dims, time_stamps)
        x = x.transpose(1, 2) 
        x = self.conv1(x)
        x = self.dropout(x)
        x = self.conv2(x)
        x = x.transpose(1, 2) # convert back to (batch_size, time_stamps, feat_dims)
        return x + res

    def transformer_classifier(self, inputs, max_time_steps):
        x = inputs # (batch_size, time_steps, feat_dims)
        for _ in range(self.num_transformer_blocks):
            x = self.transformer_encoder(x) 
        # pooling in time dimension: NSP layer
        x = torch.sum(x, dim=1) / max_time_steps
        x = self.fc_transformer(x)
        x = self.dropout(x)
        return x

    def features_to_logits(self, x, time_steps):
        """
        Args:
            x: A Tensor. Output of LSTM with shape=(batch, time_steps, self.width_lstm).
        Returns:
        - logit_stack (TensorFlow Tensor): (batch, time_steps, num_classes)
        - urgency_signals (TensorFlow Tensor): (batch, time_steps, 1)
        """
        logit_stack = []
        urgency_signals = []
        for i in range(time_steps):
            y = x[:, i, :]
            if self.is_normalize:
                y = self.layer_norm_3(y)
            # final FC layer
            y_class = self.fc_logits(y)
            logit_stack.append(y_class)

            # branch into the urgency signal
            y_urgency = self.fc_urgency(y)
            urgency_signals.append(y_urgency)
        logit_stack = torch.stack(logit_stack, dim=1)
        urgency_signals = torch.stack(urgency_signals, dim=1) 
        return logit_stack, urgency_signals

    def positional_encoding(self, inputs, max_freq=10000):
        assert len(inputs.shape) == 3 # input size: (batch_size, time_steps, feat_dims)
        batch_size = inputs.shape[0]
        maxpos = inputs.shape[-2]
        d_model = inputs.shape[-1]
        x = np.arange(0, maxpos)
        y = np.arange(0, d_model)
        xx, yy = np.meshgrid(x, y, indexing='ij')
        arg = xx/max_freq**(2*(yy//2)/d_model)
        cos_mask = y % 2
        sin_mask = 1 - cos_mask
        PE = np.sin(arg) * sin_mask + np.cos(arg) * cos_mask
        return np.repeat(PE[np.newaxis, :, :], batch_size, dim=0)

    def summarize_model(self, tb_writer):
        logger.info('\n' + pyfiglet.figlet_format(
            'SPRT-    \nTANDEM', font='banner3-D'))
        
        self.eval()
        example_input = torch.randn(self.batch_size, self.order_sprt + 1, self.feat_dims)
        with torch.no_grad():
            example_output, _ = self.forward(example_input)
            tb_writer.add_graph(self, example_input)

        # Print the model summary containing the number of trainable parameters
        num_params = sum(p.numel() for p in self.parameters() if p.requires_grad)
        logger.info(f'model summary:\n{self}')
        logger.info(f'input shape: {example_input.shape}')
        logger.info(f'output shape: {example_output.shape}')
        logger.info(f'Number of trainable parameters:, {num_params}')
    
    def forward(self, inputs):
        """Calc logits.
        Args:
            inputs: A Tensor with shape=(batch, time_steps, feature dimension). E.g. (128, 20, 784) for nosaic MNIST.
            training: A boolean. Training flag used in BatchNormalization and dropout.
        Returns:
            outputs: A Tensor with shape=(batch, time_steps, num_classes).
        """
        # inputs: (batch_size, time_steps, feat_dims)
        batch_size, time_steps, _ = inputs.shape 

        outputs_pool = []
        for i in range(time_steps):
            targinputs = inputs[:, :i+1, :]
            
            if self.is_positional_encoding: # positional encoding or not
                if self.is_trainable_encoding:
                    targinputs += self.trainable_pos_encoding[:i+1, :].unsqueeze(0).repeat(batch_size, 1, 1)
                else:
                    targinputs += self.positional_encoding(targinputs)   
            outputs_pool.append(
                self.transformer_classifier(
                    targinputs, max_time_steps=time_steps)
            )   
        outputs = torch.stack(outputs_pool, dim=1) # A Tensor with shape=(batch, time_steps, num_classes)
        # Make logits
        outputs, urgency_signal = self.features_to_logits(
            outputs,time_steps=time_steps)

        return outputs, urgency_signal
```

# Create a Base layer

```python
class BaseTANDEM(nn.Module):
    '''
    Base PyTorch model that defines utility functions that are common to all models.

    This class inherits from `nn.Module`. All TANDEM models should inherit from this class 
    instead of directly inheriting from `nn.Module` to make use of the `summarize_model` function.
    '''
    
    def __init(self):
        super(BaseTANDEM, self).__init__()
        
    def features_to_logits(self, x, time_steps):
        """
        Args:
            x: A Tensor. Output of LSTM with shape=(batch, time_steps, self.width_lstm).
        Returns:
        - logit_stack (TensorFlow Tensor): (batch, time_steps, num_classes)
        - urgency_signals (TensorFlow Tensor): (batch, time_steps, 1)
        """
        logit_stack = []
        urgency_signals = []
        for i in range(time_steps):
            y = x[:, i, :]
            if self.is_normalize:
                y = self.layer_norm_3(y)
            # final FC layer
            y_class = self.activation_logit(y)
            y_class = self.fc_logits(y)
            logit_stack.append(y_class)

        logit_stack = torch.stack(logit_stack, dim=1)
        return logit_stack

    def features_to_logits_urgent(self, x, time_steps):
        """
        Args:
            x: A Tensor. Output of LSTM with shape=(batch, time_steps, self.width_lstm).
        Returns:
        - logit_stack (TensorFlow Tensor): (batch, time_steps, num_classes)
        - urgency_signals (TensorFlow Tensor): (batch, time_steps, 1)
        """
        logit_stack = []
        urgency_signals = []
        for i in range(time_steps):
            y = x[:, i, :]
            if self.is_normalize:
                y = self.layer_norm_3(y)
            # final FC layer
            y_class = self.activation_logit(y)
            y_class = self.fc_logits(y)
            logit_stack.append(y_class)

            # branch into the urgency signal
            y_urgency = self.activation_urgency(y)
            y_urgency = self.fc_urgency(y)
            urgency_signals.append(y_urgency)
        logit_stack = torch.stack(logit_stack, dim=1)
        urgency_signals = torch.stack(urgency_signals, dim=1) 
        return logit_stack, urgency_signals

    def positional_encoding(self, inputs, max_freq=10000):
        assert len(inputs.shape) == 3 # input size: (batch_size, time_steps, feat_dims)
        batch_size = inputs.shape[0]
        maxpos = inputs.shape[-2]
        d_model = inputs.shape[-1]
        x = np.arange(0, maxpos)
        y = np.arange(0, d_model)
        xx, yy = np.meshgrid(x, y, indexing='ij')
        arg = xx/max_freq**(2*(yy//2)/d_model)
        cos_mask = y % 2
        sin_mask = 1 - cos_mask
        pos_encoding = np.sin(arg) * sin_mask + np.cos(arg) * cos_mask
        return np.repeat(pos_encoding[np.newaxis, :, :], batch_size, dim=0)  

    def summarize_model(self, tb_writer):
        logger.info('\n' + pyfiglet.figlet_format(
            'SPRT-    \nTANDEM', font='banner3-D'))

        self.eval()
        example_input = torch.randn(self.batch_size, self.order_sprt + 1, self.feat_dims)
        with torch.no_grad():
            example_output, _ = self.forward(example_input)
            tb_writer.add_graph(self, example_input)

        # Print the model summary containing the number of trainable parameters
        num_params = sum(p.numel() for p in self.parameters() if p.requires_grad)
        logger.info(f'model summary:\n{self}')
        logger.info(f'input shape: {example_input.shape}')
        logger.info(f'output shape: {example_output.shape}')
        logger.info(f'Number of trainable parameters:, {num_params}')

        

class TANDEMformer(BaseTANDEM):
    '''
     Transformer with the Normalized Summation Pooling (NSP) layer for precise SDRE.
    '''
    def __init__(self, config):
        
        super(TANDEMformer, self).__init__()

        # check if necessary parameters are defined in the config file
        requirements = set(['NUM_CLASSES', 'TIME_STEPS', 'FEAT_DIMS',
            'NUM_BLOCKS', 'HEAD_SIZE', 'NUM_HEADS', 'FF_DIM', 'MLP_UNITS',
            'ACTIVATION', 'ACTIVATION_FC', 'ACTIVATION_URGENCY',
            'IS_NORMALIZE', 'IS_POSITIONAL_ENCODING', 'IS_TRAINABLE_ENCODING', 
            'DROPOUT', 'ALPHA', 'BATCH_SIZE', 'IS_ADAPTIVE_LOSS', 'ORDER_SPRT'])
        conf = extract_params_from_config(requirements, config)

        # Parameters
        self.batch_size = conf.batch_size
        self.time_steps = conf.time_steps
        self.order_sprt = conf.order_sprt
        self.feat_dims = conf.feat_dims
        self.num_transformer_blocks = conf.num_blocks
        self.is_normalize = conf.is_normalize
        self.is_positional_encoding = conf.is_positional_encoding # positional embedding
        self.is_trainable_encoding = conf.is_trainable_encoding

        # activation functions
        self.activation = get_activation(conf.activation)
        self.activation_logit = get_activation(conf.activation_fc)
        self.activation_urgency = get_activation(conf.activation_urgency)
        
        # transformer building blocks
        self.MultiHeadAttention = nn.MultiheadAttention(
            embed_dim=self.feat_dims, num_heads=conf.num_heads, dropout=conf.dropout)
        self.conv1 = nn.Conv1d(
            in_channels=self.feat_dims, out_channels=conf.ff_dim, kernel_size=1)
        self.conv2 = nn.Conv1d(
            in_channels=conf.ff_dim, out_channels=self.feat_dims, kernel_size=1)
        if self.is_normalize:
            self.layer_norm_1 = nn.LayerNorm(self.feat_dims)
            self.layer_norm_2 = nn.LayerNorm(self.feat_dims)
            self.layer_norm_3 = nn.LayerNorm(self.feat_dims)
        self.dropout = nn.Dropout(p=conf.dropout)

        # logit computation
        self.fc_transformer = nn.Linear(self.feat_dims, conf.mlp_units)
        self.fc_urgency = nn.Linear(conf.mlp_units, 1)
        self.fc_logits = nn.Linear(conf.mlp_units, conf.num_classes)
        if self.is_trainable_encoding:
            self.trainable_pos_encoding = nn.Parameter(
                torch.randn((self.order_sprt + 1, self.feat_dims)) * 0.1)

        if conf.is_adaptive_loss:
            # adaptive loss weights for mce, lllr, ausat, and weight decay
            self.adaptive_loss_weights = nn.Parameter(torch.ones(4))
            
    # transformer block 
    def transformer_encoder(self, inputs):
        # self-attention 
        x = inputs # (batch_size, time_steps, feat_dims)
        if self.is_normalize:
            x = self.layer_norm_1(x)
        # MultiHeadAttention returns the tuple of two tensors, 
        # output of multi-head attention and attention weights.
        x = self.MultiHeadAttention(x, x, x)[0]
        x = self.dropout(x)
        res = x + inputs

        # feedforward
        if self.is_normalize:
            x = self.layer_norm_2(res)
        else:
            x = res
        # PyTorch conv1d expect an input shape to be (batch_size, feat_dims, time_stamps)
        x = x.transpose(1, 2) 
        x = self.conv1(x)
        x = self.dropout(x)
        x = self.conv2(x)
        x = x.transpose(1, 2) # convert back to (batch_size, time_stamps, feat_dims)
        return x + res

    def transformer_classifier(self, inputs, max_time_steps):
        x = inputs # (batch_size, time_steps, feat_dims)
        for _ in range(self.num_transformer_blocks):
            x = self.transformer_encoder(x) 
        # pooling in time dimension: NSP layer
        x = torch.sum(x, dim=1) / max_time_steps
        x = self.fc_transformer(x)
        x = self.dropout(x)
        return x
    
    def forward(self, inputs):
        """Calc logits.
        Args:
            inputs: A Tensor with shape=(batch, time_steps, feature dimension). E.g. (128, 20, 784) for nosaic MNIST.
            training: A boolean. Training flag used in BatchNormalization and dropout.
        Returns:
            outputs: A Tensor with shape=(batch, time_steps, num_classes).
        """
        # inputs: (batch_size, time_steps, feat_dims)
        batch_size, time_steps, _ = inputs.shape 

        outputs_pool = []
        for i in range(time_steps):
            targinputs = inputs[:, :i+1, :]
            
            if self.is_positional_encoding: # positional encoding or not
                if self.is_trainable_encoding:
                    targinputs += self.trainable_pos_encoding[:i+1, :].unsqueeze(0).repeat(batch_size, 1, 1)
                else:
                    targinputs += self.positional_encoding(targinputs)   
            outputs_pool.append(
                self.transformer_classifier(
                    targinputs, max_time_steps=time_steps)
            )   
        outputs = torch.stack(outputs_pool, dim=1) # A Tensor with shape=(batch, time_steps, num_classes)
        # Make logits
        outputs, urgency_signal = self.features_to_logits_urgent(
            outputs,time_steps=time_steps)

        return outputs, urgency_signal
    

    
class B2BsqrtTANDEM(BaseTANDEM):

    def __init__(self, config):
        
        super(B2BsqrtTANDEM, self).__init__()
        
        # check if necessary parameters are defined in the config file
        requirements = set(['NUM_CLASSES', 'WIDTH_LSTM', 'TIME_STEPS', 'FEAT_DIMS',
            'ACTIVATION', 'ACTIVATION_RECURRENT', 'ACTIVATION_FC', 'ACTIVATION_URGENCY',
            'IS_NORMALIZE', 'IS_POSITIONAL_ENCODING', 'IS_TRAINABLE_ENCODING', 
            'DROPOUT', 'ALPHA', 'BATCH_SIZE', 'IS_ADAPTIVE_LOSS', 'ORDER_SPRT'])
        conf = extract_params_from_config(requirements, config)

        # Parameters
        self.batch_size = conf.batch_size
        self.feat_dims = conf.feat_dims
        self.time_steps = conf.time_steps
        self.order_sprt = conf.order_sprt
        self.num_classes = conf.num_classes
        self.width_lstm = conf.width_lstm
        self.dropout = conf.dropout
        self.is_normalize = conf.is_normalize
        if self.is_normalize:
            self.layer_norm = nn.LayerNorm(
                normalized_shape=[self.feat_dims, self.time_steps]) 
        self.is_positional_encoding = conf.is_positional_encoding
        self.is_trainable_encoding = conf.is_trainable_encoding
        if self.is_trainable_encoding:
            self.trainable_pos_encoding = nn.Parameter(
                torch.randn(self.order_sprt + 1, self.feat_dims))

        # activation functions
        self.activation = get_activation(conf.activation)
        self.recurrent_activation = get_activation(conf.activation_recurrent)
        self.activation_logit = get_activation(conf.activation_fc)
        self.activation_urgency = get_activation(conf.activation_urgency)
        
        logger.warning('Currently B2Bsqrt is not supported.')
        self.rnn = nn.LSTM(
            input_size=self.feat_dims,
            hidden_size=self.width_lstm,
            num_layers=1,
            batch_first=True,
)
        # classification head: logits and urgency signals
        self.fc_logits = nn.Linear(in_features=self.width_lstm, out_features=self.num_classes, bias=True)
        self.fc_urgency = nn.Linear(in_features=self.width_lstm, out_features=1, bias=True)
        
        if conf.is_adaptive_loss:
            # adaptive loss weights
            self.adaptive_loss_weights = nn.Parameter(torch.randn(4))

    def forward(self, inputs):
        """Calc logits.
        Args:
            inputs: A Tensor with shape=(batch, time_steps, feature dimension). E.g. (128, 20, 784) for nosaic MNIST.
            training: A boolean. Training flag used in BatchNormalization and dropout.
        Returns:
            outputs: A Tensor with shape=(batch, time_steps, num_classes).
        """
        # inputs: (batch_size, time_steps, feat_dims)
        batch_size, time_steps, _ = inputs.shape 
        
        # Positinal encoding
        if self.is_positional_encoding:
            if self.is_trainable_encoding:
                inputs += self.trainable_pos_encoding.unsqueeze(0).repeat(batch_size, 1, 1)
            else:
                # (batch_size, time_steps, feat_dims)
                inputs += self.positional_encoding(inputs) 
        # Feedforward
        outputs, _ = self.rnn(inputs)

        # Make logits
        outputs, urgency_signal = self.features_to_logits_urgent(
            outputs, time_steps=time_steps)

        return outputs, urgency_signal # A Tensor with shape=(batch, order_sprt+1, num_classes)
```

# Positional Encoding from PyTorch official

```python
import math
import os
from tempfile import TemporaryDirectory
from typing import Tuple

import torch
from torch import nn, Tensor
import torch.nn.functional as F
from torch.nn import TransformerEncoder, TransformerEncoderLayer
from torch.utils.data import dataset


class PositionalEncoding(nn.Module):

    def __init__(self, feat_dim: int, dropout: float = 0.1, max_len: int = 5000):
        super().__init__()
        self.dropout = nn.Dropout(p=dropout)
        self.feat_dim = feat_dim
        position = torch.arange(max_len).unsqueeze(1)
        div_term = torch.exp(torch.arange(0, feat_dim, 2) * (-math.log(10000.0) / feat_dim))
        pe = torch.zeros(max_len, 1, feat_dim)
        pe[:, 0, 0::2] = torch.sin(position * div_term)
        pe[:, 0, 1::2] = torch.cos(position * div_term)
        # register the pe tensor as a buffer, which means that it will be included in the model's 
        # state dictionary but won't be considered as a model parameter that needs to be optimized 
        # during training. 
        self.register_buffer('pe', pe)

    def forward(self, x: Tensor, batch_first: bool = True) -> Tensor:
        """
        Args:
            x: Tensor, shape [time_steps, batch_size, embedding_dim] if batch_first=False
                   else [batch_size, time_steps, embedding_dim] if batch_first=True 
        """
        assert x.shape[-1] == self.feat_dim
        assert len(x.shape) == 3, 'Expecting a Tensor with dimenstions batch_size, time_steps, and embedding_dim.'
        x = x.permute(1, 0, 2) if batch_first else x
        x = x + self.pe[:x.size(0)]
        x = self.dropout(x)
        return x.permute(1, 0, 2) if batch_first else x
```

```python
class TransformerModel(nn.Module):

    def __init__(self, ntoken: int, d_model: int, nhead: int, d_hid: int,
                 nlayers: int, dropout: float = 0.5):
        super().__init__()
        
        self.model_type = 'Transformer'
        self.layer_norm = nn.LayerNorm(d_model)
        
        self.pos_encoder = PositionalEncoding(d_model, dropout)
        
        encoder_layers = TransformerEncoderLayer(
            d_model, nhead, dim_feedforward=d_hid, dropout=dropout,
            activation='relu', batch_first=True, norm_first=False)
        
        self.transformer_encoder = TransformerEncoder(encoder_layers, nlayers)
#         self.encoder = nn.Embedding(ntoken, d_model)
        self.d_model = d_model
        self.decoder = nn.Linear(d_model, d_model)
    
        self.init_weights()

    def init_weights(self) -> None:
        initrange = 0.1
#         self.encoder.weight.data.uniform_(-initrange, initrange)
        self.decoder.bias.data.zero_()
        self.decoder.weight.data.uniform_(-initrange, initrange)

    def forward(self, src: Tensor) -> Tensor:
        """
        Args:
            src: Tensor, shape [seq_len, batch_size]
            src_mask: Tensor, shape [seq_len, seq_len]

        Returns:
            output Tensor of shape [seq_len, batch_size, ntoken]
        """
#         src = self.encoder(src) * math.sqrt(self.d_model)
        src = self.pos_encoder(src)
        output = self.transformer_encoder(src)
        print(f'{output.shape=}')
        output = self.decoder(output)
        print(f'{output.shape=}')
        return output

model = TransformerModel(ntoken=50, d_model=128, nhead=2, d_hid=64, nlayers=2, dropout=0.0)
device = torch.device('cuda')
model.to(device)
model = torch.compile(model)
print(model)
```

```python
class BaseTANDEM(nn.Module):
    '''
    Base PyTorch model that defines utility functions that are common to all models.

    This class inherits from `nn.Module`. All TANDEM models should inherit from this class 
    instead of directly inheriting from `nn.Module` to make use of the `summarize_model` function.
    '''
    
    def __init__(self):
        super().__init__()
        
    def features_to_logits(self, x, time_steps):
        '''
        Args:
            x: A Tensor. Output of LSTM with shape=(batch, time_steps, self.width_lstm).
        Returns:
        - logit_stack (TensorFlow Tensor): (batch, time_steps, num_classes)
        - urgency_signals (TensorFlow Tensor): (batch, time_steps, 1)
        '''
        logit_stack = []
        urgency_signals = []
        for i in range(time_steps):
            y = x[:, i, :]
            if self.is_normalize:
                y = self.layer_norm_3(y)
            # final FC layer
            y_class = self.activation_logit(y)
            y_class = self.fc_logits(y)
            logit_stack.append(y_class)

        logit_stack = torch.stack(logit_stack, dim=1)
        return logit_stack

    def features_to_logits_urgent(self, x, time_steps):
        '''
        Args:
            x: A Tensor. Output of LSTM with shape=(batch, time_steps, self.width_lstm).
        Returns:
        - logit_stack (TensorFlow Tensor): (batch, time_steps, num_classes)
        - urgency_signals (TensorFlow Tensor): (batch, time_steps, 1)
        '''
        logit_stack = []
        urgency_signals = []
        for i in range(time_steps):
            y = x[:, i, :]
            if self.is_normalize:
                y = self.layer_norm_3(y)
            # final FC layer
            y_class = self.activation_logit(y)
            y_class = self.fc_logits(y)
            logit_stack.append(y_class)

            # branch into the urgency signal
            y_urgency = self.activation_urgency(y)
            y_urgency = self.fc_urgency(y)
            urgency_signals.append(y_urgency)
        logit_stack = torch.stack(logit_stack, dim=1)
        urgency_signals = torch.stack(urgency_signals, dim=1) 
        return logit_stack, urgency_signals

    def positional_encoding(self, inputs, max_freq=10000):
        assert len(inputs.shape) == 3 # input size: (batch_size, time_steps, feat_dim)
        batch_size = inputs.shape[0]
        maxpos = inputs.shape[-2]
        d_model = inputs.shape[-1]
        x = np.arange(0, maxpos)
        y = np.arange(0, d_model)
        xx, yy = np.meshgrid(x, y, indexing='ij')
        arg = xx/max_freq**(2*(yy//2)/d_model)
        cos_mask = y % 2
        sin_mask = 1 - cos_mask
        pos_encoding = np.sin(arg) * sin_mask + np.cos(arg) * cos_mask
        return np.repeat(pos_encoding[np.newaxis, :, :], batch_size, axis=0)  

    def summarize_model(self, tb_writer):
        logger.info('\n' + pyfiglet.figlet_format(
            'SPRT-    \nTANDEM', font='banner3-D'))
        logger.info(colored('official PyTorch implementation.', 'blue'))
        self.eval()
        example_input = torch.randn(self.batch_size, self.order_sprt + 1, self.feat_dim)
        with torch.no_grad():
            example_output, _ = self.forward(example_input)
            tb_writer.add_graph(self, example_input)

        # Print the model summary containing the number of trainable parameters
        num_params = sum(p.numel() for p in self.parameters() if p.requires_grad)
        logger.info(f'model summary:\n{self}')
        logger.info(f'input shape: {example_input.shape}')
        logger.info(f'output shape: {example_output.shape}')
        logger.info(f'Number of trainable parameters:, {num_params}')

        
class TANDEMproto(BaseTANDEM):

    def __init__(self, ntoken: int, d_model: int, nhead: int, d_hid: int,
                 nlayers: int, dropout: float = 0.5):
        super().__init__()
        self.model_type = 'Transformer'
        self.pos_encoder = PositionalEncoding(d_model, dropout)
        
        encoder_layers = TransformerEncoderLayer(
            d_model, nhead, dim_feedforward=d_hid, dropout=dropout,
            activation='relu', batch_first=True, norm_first=False)
        
        self.transformer_encoder = TransformerEncoder(encoder_layers, nlayers)
        self.d_model = d_model
        self.decoder = nn.Linear(d_model, d_model)
        self.dropout = nn.Dropout(0.1)

        self.is_normalize = True
        
        self.init_weights()

    def init_weights(self) -> None:
        initrange = 0.1
        self.decoder.bias.data.zero_()
        self.decoder.weight.data.uniform_(-initrange, initrange)
    
    def normalized_sum_pool(self, x: Tensor, max_time_steps: int) -> Tensor:
        '''
        NSP layer for precise sequential density ratio estimation.
        (Ebihara+, ICASSP 2023)
        
        Args:
        -x (Tensor): size (effective_batch_size, effective_time_steps, feat_dim)
         effective_time_steps \in {1, max_time_steps}
        -max_time_steps (int): maximum duration to consider. e.g) order_sprt + 1.
        
        Return:
        torch.Tensor of the size size as x.
        '''
        return torch.sum(x, dim=1) / max_time_steps
    
    def transformer_classifier(self, inputs, max_time_steps):
        # pooling in time dimension: NSP layer
        inputs = self.normalized_sum_pool(inputs, max_time_steps)
        inputs = self.decoder(inputs)
        inputs = self.dropout(inputs)
        return inputs
    
    def forward(self, inputs: Tensor) -> Tensor:
        """
        Args:
            src: Tensor, shape [seq_len, batch_size]
            src_mask: Tensor, shape [seq_len, seq_len]

        Returns:
            output Tensor of shape [seq_len, batch_size, ntoken]
        """
        batch_size, time_steps, _ = inputs.shape 

        outputs_pool = []
        for i in range(time_steps):
            targinputs = inputs[:, :i+1, :]
            
            # add positional encoding, either trainable or constant
            targinputs = self.pos_encoder(targinputs, batch_first=True)
            targoutputs = self.transformer_encoder(targinputs)
            
            outputs_pool.append(
                self.transformer_classifier(targoutputs, max_time_steps=time_steps)
            )   
        outputs = torch.stack(outputs_pool, dim=1) # (batch_size, time_steps, mlp_feats)
        outputs,  = self.features_to_logits(outputs, time_steps=time_steps) # (batch_size, time_steps, num_classes)

        return outputs

model = TANDEMproto(ntoken=50, d_model=128, nhead=2, d_hid=64, nlayers=2, dropout=0.0)
device = torch.device('cuda')
model.to(device)
# model = torch.compile(model)
print(model)
```
